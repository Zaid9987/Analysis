```python
# Visualization

df = pd.read_csv("Automobile_data.csv")
df
```

```python
df["engine-size"].hist(color="red") # Univeriate analysis: used to work on single column. hist is used to plot histogram type.
# Engine-size ranges from 50-400.
# Most of the engine are of range 90-100.
# Some engine have the engine size of 300+
```

```python
df[["price","height"]].hist(figsize=(10,5),color="red") #we can also adjust the graph color.
plt.tight_layout()  #used to automatically adjust subplot parameters.
plt.show()          #used to display the figures.
```

```python
# line-Plot is used to draw line chart.

x = np.arange(10)
y = np.sqrt(x)
sns.lineplot(x,y)
plt.grid(True)
plt.show()
```

```python
# Dist-Plot is used to access data in histogram and line chart.

sns.distplot(df['price'])
plt.show()
```

```python
mprice=df['price'].mean() 
meprice=df['price'].median() # used to plot user's mean and median on the graph.

sns.distplot(df['price'])
plt.axvline(mprice,color='red')  
plt.axvline(meprice,color='blue')
plt.grid(True) # grid is used to show vertical and horizontal lines on graph.
plt.show()
```

```python
# Count-Plot: used to show the counts of observation.

sns.countplot(df['body-style'])
plt.show()
# Here, sedan is mostly saled body-style vehicle with 80+ sales.
# Hatchback is 2nd most saled vehicle bodytype with 60+ sales.
# Wagon on 3rd position which sales between 20-40.
# Hardtop sales between 8-20, followed by convertible.
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df,y='make') #y='make' plot observation on y-axis
plt.tight_layout()
plt.grid()
plt.show()
''' Toyota is most selled car topping on no.1 with more tha 30+ sales, 
    Nissan is on 2nd position,
    and Mazda is on 3rd position.
'''
# lowest selled cars range between 0-5 are 1.Mercury, 2.Renault, 3.Chevrolet/alfa-romero/jaguar, 4.isuzu
```

```python
plt.figure(figsize=(5,5))
sns.countplot(df['fuel-type'])
plt.tight_layout()
plt.grid()
plt.show()

# Most cars run on gas with a number of 175+
# Cars running on diesel are less then 25.
```

```python
sns.countplot(data=df,x='body-style',hue='fuel-type') # hue is use to plot 2 columns in a graph.
plt.grid()
plt.show()
# here, we can see sedan uses most of the diesel and gas other than any body-style vehical.
# convertible body-style uses only gas.
```

```python
# sub-plot: is used to divide and add multiple columns in a graph.

plt.figure(1)
plt.subplot(2,2,1)
df['engine-type'].value_counts().plot(figsize=(10,8),kind='bar',color='red')
plt.title("Engine Type Frequence Diagram")
plt.ylabel("No. of Engine-Type")
plt.xlabel("Engine Type")
plt.grid()

plt.subplot(2,2,2)
df['drive-wheels'].value_counts().plot(figsize=(10,8),kind='bar',color='blue')
plt.title("Drive-Wheels Type Frequence Diagram")
plt.ylabel("No. of Drive-Wheels Type")
plt.xlabel("Drive-Wheels Type")
plt.grid()

plt.subplot(2,2,3)
df['fuel-type'].value_counts().plot(figsize=(10,8),kind='bar',color='green')
plt.title("Fuel-Type Frequence Diagram")
plt.ylabel("No. of Fuel-Type")
plt.xlabel("Fuel-Type")
plt.grid()

plt.subplot(2,2,4)
df['body-style'].value_counts().plot(figsize=(10,8),kind='bar',color='purple')
plt.title("Body-Style Type Frequence Diagram") # to give title or heading of a graph
plt.ylabel("No. of Body-Style")                # to give label on y-axis
plt.xlabel("Body-Style Type")                  #to give label on x-axis
plt.grid()   

plt.tight_layout()
plt.show()
```

```python
#pie-chart is a circular representation on the data

explode=[0,0,0,0,0.1] # explode parameter is used to access particular column or data in chart.
df['body-style'].value_counts().plot(figsize=(6,6),kind='pie',autopct='%.2f%%',explode=explode)
# autopct="% %" is used to access pie area.

plt.title("Body-Style Pie-Chart")
plt.xlabel("Type of Vehicals")
plt.ylabel("Body-Style Type")
```

```python
#Bivariate analysis: analysis of any concurrent relation between two variables or attributes.

df.corr() #shows co-relation between different columns
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True,cmap='Blues') # annot=True to access values in plot
plt.show()
#the Heatmap in Seaborn is to display a color-coded correlation matrix for easy visualization of the relationship between data.

'''The light sides of the color is positively corelated
and the darker side of the plot is negatively corelated'''

'''City-mpg and high-way mpg is highly corelated
engine-size and high-way mpg is less corelated'''
```

```python
#Box-plot or Whiskers-plot: Give information of an out-lier

plt.figure(figsize=(5,10))
sns.boxplot(data=df,x='drive-wheels',y='price')
plt.show()
#There are out-liers in the data-set.

'''Mercedes benz, BMW, Jaguar produces expensive cars starting from 25k.
   Honda, Nissan, Toyota,etc produces budget models with lower prices. 
   Most of the companies produce cars in the range below 25k. 
   Hard-top model are expensive followed by convertable. 
   Rear-Wheel-Drive vehicals are expensive then Front-Wheel-Drive'''
```

```python
#scatterplot : gives data in dot or point format.

plt.figure(figsize=(10,8))
sns.scatterplot(data=df,x='highway-mpg',y='price')
plt.show()
```

```python
#Pair-plot: combination of scatter plots and histograms.

plt.figure(figsize=(5,5))
sns.pairplot(df[['city-mpg','engine-size','fuel-type','price']],diag_kind='hist')
plt.show()
```
